```python
df = pd.read_csv('US Minimum Wage Data.csv',encoding='cp1252')
```

```python
df.head()
```

Dropping footnotes since we won't be needing it, and it contains many null objects

```python
df = df.drop(['Footnote'], axis=1)
```

```python
df.head()
```

```python
df.isnull()
```

Creating pivot table to calculate growth rate of minimum wage by state using inflated $

```python
state_in = df.pivot(index='Year', columns='State', values='State.Minimum.Wage.2020.Dollars')
```

```python
state_in.head()
```

Calculating the growth rate of each state's minumum wage using 2020 $--- growth rate calc =

((New-old)/old)*100

let's try this formula using Alaska...

```python
(state_in['Alaska'].iloc[-1] - state_in['Alaska'].iloc[0]) / state_in['Alaska'].iloc[0] * 100
```

Function for growth rate calculation

```python
def rate(x):
  rate  = (state_in.iloc[-1] - state_in.iloc[0]) / state_in.iloc[0] * 100
  return rate
```

Transposing to make the output readable

```python
sin_t = state_in.transpose()
```

```python
sin_t.head()
```

```python
in_gr = sin_t.apply(rate)
```

```python
in_gr.head()
```

Extracting one column to create a new data frame

```python
df_gr = in_gr[1968]
```

```python
df_gr.head()
```

```python
gr = pd.DataFrame(df_gr)
```

```python
gr.head()
```

Dropping both countries that do not have a state minimum wage and countries that did not start using minimum wages early enough to fit into our 52 year window of analysis


```python
gr = gr.replace([np.inf, -np.inf], np.nan)
```

```python
gr.head()
```

```python
gr = gr.dropna(how='any')
```

The growth rate for each US state's minimum wage in 2020 $ (1968-2020) is...
